```python
#logistic regression (spam or notspam)

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score

data = pd.DataFrame({
    'message': [
        "Win money now", "Hello friend", "Free offer!!!",
        "Let's meet tomorrow", "Claim your prize", "How are you?"
    ],
    'label': [1, 0, 1, 0, 1, 0]  })

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['message'])
y = data['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Spam Detection Accuracy:", accuracy_score(y_test, y_pred))

msg = ["Free money offer"]
msg_vec = vectorizer.transform(msg)
print("Prediction (1=Spam, 0=Not Spam):", model.predict(msg_vec)[0])
```
