```python
# Bubble sort
n = int(input("Enter number of elements: "))
arr = []

for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

print("\nOriginal array:", arr)

#bubble sort logic
for i in range(n-1):
  for j in range (n-1-1):
    if arr[j] > arr[j+1]:
      arr[j], arr[j+1]= arr[j+1], arr[j]
print("sorted array:", arr)
```
