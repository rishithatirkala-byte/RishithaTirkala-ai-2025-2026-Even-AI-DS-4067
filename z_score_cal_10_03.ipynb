```python
#z-score
import numpy as np

data = [10, 20, 30, 40, 50]

mean = np.mean(data)
std_dev = np.std(data, ddof=1)

print("Mean:", mean)
print("Standard deviation:", std_dev)

# Z-score of 40
x_value = 40
z_score_40 = (x_value - mean) / std_dev
print("Z-score of 40:", z_score_40)

# Z-scores of all data
z_scores = [(x - mean) / std_dev for x in data]
print("Z-scores of data:", z_scores)
```
