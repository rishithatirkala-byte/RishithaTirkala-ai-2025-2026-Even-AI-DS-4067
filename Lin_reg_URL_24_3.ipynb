```python
#Linear regression with an example URL
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

url = "https://raw.githubusercontent.com/ybiFoundation/Dataset/main/Salary%20Data.csv"

# load dataset
data = pd.read_csv(url)
print("Dataset preview")
print(data.head())

x = data[['Experience Years']]
y = data['Salary']

model = LinearRegression()
model.fit(x, y)

print("\n model parameters")
print("slope:", model.coef_[0])
print("intercept:", model.intercept_)

years = [[5]]
predicted_salary = model.predict(years)

print("\n predicted salary for 5 years experience:")
print(predicted_salary[0])

y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)

print("\n mean squared error:", mse)

plt.plot(x, y_pred)
plt.scatter(x, y)
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.title("Linear Regression salary prediction")
plt.show()
```
