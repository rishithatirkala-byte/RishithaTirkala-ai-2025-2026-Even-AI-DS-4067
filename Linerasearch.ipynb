```python
#linear search
n = int(input("Enter number of elements: "))
arr = []

# input elements
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

key = int(input("Enter element to search: "))

found = False

# linear search
for i in range(n):
    if arr[i] == key:
        print(f"{key} found at index {i}")
        found = True
        break

if not found:
    print("Element not found")
```
