```python
#Pandas Dataframe creation and indexing using iloc in python
import pandas as pd

data = {
    "ID": [101, 102, 103],
    "Name": ['abc', 'xyz', 'pqr']
}

df = pd.DataFrame(data)

print(df)
print(df[["Name"]])
print("location 0\n", df.iloc[0])
print("location to 3\n", df.iloc[0:3, 1:2])
```
