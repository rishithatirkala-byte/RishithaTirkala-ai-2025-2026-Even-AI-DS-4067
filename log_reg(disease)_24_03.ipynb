```python
# Logistic regression(Disease or not disease)

import pandas as pd
from sklearn.linear_model import LogisticRegression

# Sample dataset
data = pd.DataFrame({
    'age': [25, 45, 35, 50, 23, 40],
    'bp': [120, 140, 130, 150, 110, 135],
    'disease': [0, 1, 0, 1, 0, 1]  # 1 = Disease
})

X = data[['age', 'bp']]
y = data['disease']

# Model
model = LogisticRegression()
model.fit(X, y)

# Predict for new patient
new_patient = [[30, 125]]
prediction = model.predict(new_patient)

print("Disease Prediction (1=Yes, 0=No):", prediction[0])
```
