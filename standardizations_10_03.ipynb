```python
#standardizations
import numpy as np

# Data values
data = np.array([10,20,30,40,50,60,70])

# Mean of data
mean = np.mean(data)
print("Mean:", mean)

# Standard deviation
std = np.std(data)
print("Standard Deviation:", std)

# Standardization (Z-score)
z_scores = (data - mean) / std

print("\nStandardized Values (Z-scores):")
for i in range(len(data)):
    print("Value:", data[i], " Z-score:", z_scores[i])
```
